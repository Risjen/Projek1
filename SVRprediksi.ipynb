```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
##import io
```

```python
#membaca data harga cabai pada csv
data=pd.read_csv('data_harga_cabe.csv')
```

```python
cdf=data[['tahun 2019','tahun 2020','tahun 2021','tahun 2022']]
```

```python
#melihat plot data harga cabai merah setiap tahun dari 2019-2022
cdf.plot()
```

```python
#membagi data menjadi data input dan data target 2019-2021(x) adalah data input dan 2022(y) merupakan target
x=cdf.iloc[:,:-1].values
y=cdf.iloc[:,-1:].values
```

```python
#membagi data menjadi data training dan data latih 
from sklearn.model_selection import train_test_split
xtrain,xtes,ytrain,ytes=train_test_split(x,y,train_size=0.4,random_state=45)
```

```python
#mencari svr terbaik dengan menggunakan gridsearch
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
my_param_grid={'C':[10000,20000,30000,40500],'epsilon':[300,400,500],'kernel':['rbf']}
GridSearchCV(estimator=SVR(),param_grid=my_param_grid,refit= True ,verbose=2,cv=10)
grid_rbf=GridSearchCV(estimator=SVR(),param_grid=my_param_grid,refit= True ,verbose=2,cv=10)
```

```python
#fit digunakan untuk training data
grid_rbf.fit(xtrain,ytrain)
```

```python
grid_rbf.best_params_
```

```python
grid_rbf.best_estimator_
```

```python
ypred_rbf_optimized=grid_rbf.predict(xtes)
```

```python
#menampilkan data tes,target dan prediksi
for i in range(len(ytes)):
    print('Data:', xtes[i], '  Target:', ytes[i], '  Prediksi:', ypred_rbf_optimized[i])
```

```python
import sklearn.metrics as sm
print("MAPE = ", round(sm.mean_absolute_percentage_error(ytes,ypred_rbf_optimized),2))
```

```python
plt.figure(figsize=(10,5))
plt.plot(ytes,label='harga actual',color='blue')
plt.plot(ypred_rbf_optimized,label='harga prediksi',color='red')
plt.xlabel('actual',fontsize=10)
plt.ylabel('prediksi',fontsize=10)
plt.title('Harga Actual VS Harga Prediksi',fontsize=12)
plt.legend()
plt.show()
```

```python
#x1=13500,x2=12000,x3=18000
baru=[[13500,12000,18000]]
hasil=grid_rbf.predict(baru)
print("Hasil prediksi harga cabai merah",hasil)
```

1.alpha SVR adalah faktor pembobotan yang mengatur tingkat penalti yang diberikan kepada data yang jatuh di luar batas margin atau yang melanggar batasan kesalahan toleransi (epsilon).Dalam SVR, tujuan utama adalah untuk mencari model regresi yang dapat menghasilkan fungsi hipotesis yang paling dekat dengan data pelatihan. Dalam mencapai tujuan ini, alpha membantu mengontrol kompleksitas model dan tingkat toleransi kesalahan. Nilai alpha yang lebih besar menunjukkan penalti yang lebih tinggi terhadap pelanggaran batasan margin atau epsilon, yang mengarah pada model yang lebih sederhana dengan lebih sedikit data pelatihan yang diperhitungkan.Dengan mengatur nilai alpha, Anda dapat mengontrol trade-off antara kesalahan pelatihan dan kompleksitas model. Nilai alpha yang tinggi cenderung menghasilkan model yang lebih sederhana dengan kesalahan pelatihan yang lebih besar, sementara nilai alpha yang rendah dapat menghasilkan model yang lebih kompleks dengan kesalahan pelatihan yang lebih kecil.Dalam praktiknya, nilai alpha biasanya ditentukan melalui proses validasi silang (cross-validation) atau optimasi menggunakan teknik seperti grid search

2.Support vector pada metode SVR didapatkan melalui proses optimasi yang dilakukan selama pelatihan model. Support vector merupakan subset dari data pelatihan yang paling penting dalam membangun model regresi. Mereka adalah titik data yang jatuh di sekitar batas margin atau melanggar batasan kesalahan toleransi (epsilon).Proses optimasi dalam SVR bertujuan untuk menemukan hyperplane (bidang pemisah) yang memiliki margin maksimum dan dapat memisahkan support vector dari kelas yang berbeda. Hyperplane ini digunakan untuk membangun fungsi hipotesis yang dapat digunakan untuk melakukan prediksi pada data yang tidak terlihat sebelumnya.Dalam SVR, konsep support vector diperluas untuk regresi dengan memperkenalkan dua batas margin, yaitu batas margin atas (upper margin) dan batas margin bawah (lower margin). Support vector adalah data pelatihan yang jatuh di luar batas margin atau melanggar batasan kesalahan toleransi. Mereka memiliki kontribusi penting dalam menentukan fungsi hipotesis dan membentuk model regresi.
